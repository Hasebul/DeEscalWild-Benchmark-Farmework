```python
import glob
import os
import shutil
from pathlib import Path

def get_row_count(file_path):
    """
    Counts the number of lines (rows) in a file.
    Returns 0 if the file cannot be read.
    """
    try:
        # We open with errors='ignore' to prevent crashes on weird characters
        with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
            # Simple, memory-efficient way to count lines
            return sum(1 for _ in f)
    except Exception:
        return 0

def main():
    # 1. Setup folders and prefixes
    # Keys matches your folder names, Values matches the prefix you want to add
    folder_prefix_map = {
        "../output/gemini_2.5_flash_output/csv": "youtube_",
    }
    
    dest_dir = "data"

    # 2. Create destination folder if it doesn't exist
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)

    # 3. Initialize counters
    total_csv_found = 0
    blank_files_count = 0  # Files with < 8 rows
    transferred_files_count = 0

    print(f"Starting process... transferring files to '{dest_dir}/'\n")
    ignored_files = []

    # 4. Loop through each source folder defined in our map
    for source_folder, prefix in folder_prefix_map.items():
        
        # Check if the source folder actually exists before trying to search it
        if not os.path.exists(source_folder):
            print(f"Warning: Source folder '{source_folder}' not found. Skipping.")
            continue

        # Find all CSV files recursively
        csv_files = glob.glob(os.path.join(source_folder, '**', '*.csv'), recursive=True)
        
        for file_path in csv_files:
            total_csv_found += 1
            
            # 5. Check row count (Blank check)
            row_count = get_row_count(file_path)
            
            if row_count < 8:
                # print(f"[BLANK] Ignored: {file_path} ({row_count} rows)")
                blank_files_count += 1
                # Extract the stem (the filename without the extension)
                numerical_id = int(Path(file_path).stem)
                # print(f"[BLANK] Ignored: {file_path} ({row_count} rows) - ID: {numerical_id}")
                ignored_files.append(numerical_id)
            # else:
            #     try:
            #         # 6. Prepare new filename with prefix
            #         original_filename = os.path.basename(file_path)
            #         new_filename = f"{prefix}{original_filename}"
                    
            #         dest_path = os.path.join(dest_dir, new_filename)
                    
            #         # 7. Transfer (Copy)
            #         shutil.copy(file_path, dest_path)
            #         transferred_files_count += 1
            #     except Exception as e:
            #         print(f"[ERROR] Could not copy {file_path}: {e}")

    # 8. Final Statistics
    print("-" * 30)
    print("SUMMARY")
    print("-" * 30)
    print(f"Total CSV files found:    {total_csv_found}")
    print(f"Ignored (< 8 rows):       {blank_files_count}")
    print(f"Successfully transferred: {transferred_files_count}")
    print("-" * 30)
    ignored_files.sort()  # Sort the list of ignored file IDs
    print(ignored_files)

if __name__ == "__main__":
    main()
```
